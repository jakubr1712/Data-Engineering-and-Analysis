**Zadanie 3**
Przetestuj różne napisy formatujące. Narysuj wykres funkcji $f(x) = \sin(x)$ przy użyciu:
* zielonych linii typu "kropka-kreska";
* niebieskich trójkątów;
* czarnych linii przerywanych oraz gwiazdek.

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, axs = plt.subplots(3, 1, figsize=(8, 10))

axs[0].plot(x, y, 'g-.', label="sin(x)")
axs[0].set_title('Zielona kropka-kreska')
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('$f(x)$')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(x, y, 'b^', label="sin(x)")
axs[1].set_title('Niebieskie trójkąty')
axs[1].set_xlabel('$x$')
axs[1].set_ylabel('$f(x)$')
axs[1].grid(True)
axs[1].legend()

axs[2].plot(x, y, 'k--', label='Czarna linia przerywana')
axs[2].plot(x, y, 'k*', label='Gwiazdy')
axs[2].set_title('Czarna linia przerywana i gwiazdy')
axs[2].set_xlabel('$x$')
axs[2].set_ylabel('$f(x)$')
axs[2].grid(True)
axs[2].legend()

plt.tight_layout()
plt.show()
```

**Zadanie 5**
Narysuj wykres funkcji $\tan(x)$  na przedziale $[-2\pi, 2\pi]$. Zamieść na osi etykiety postaci $\dots,-\pi, -\frac{\pi}{2}, 0,\frac{\pi}{2}, \pi, \dots$. Umieść legendę o białym tle w lewym dolnym rogu. Zadbaj o brak "asymptot" pionowych.

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

y = np.tan(x)

y[np.abs(y) > 12] = np.nan

plt.plot(x, y, label=r'$\tan(x)$')

plt.xticks([-2*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-\pi$', r'$-\frac{\pi}{2}$', r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$2\pi$'])

plt.ylim(-10, 10)

plt.legend(loc='lower left', facecolor='white')

plt.title(r'Wykres funkcji $\tan(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'$\tan(x)$')
plt.grid(True)
plt.show()
```

Sumuj wszystkie odpowiedzi dla każdego pytania i przedstaw je w następujący sposób: dla osi Y ustaw numery pytań a dla osi X sumę odpowiedzi na każde pytanie. Zaznacz innym kolorem niż domyślny dwie najlepsze i dwa najgorsze wyniki dla przedstawionych pytań. Pamiętaj aby odpowiednio opisać wykres.

```python
with open('answers.txt', 'r') as file:
    rows = [list(map(int, line.strip().split(','))) for line in file]

sums = [sum(row[1:]) for row in rows]
questions = [row[0] for row in rows]

sorted_indices = sorted(range(len(sums)), key=lambda i: sums[i])
highlight_indices = sorted_indices[:2] + sorted_indices[-2:]

plt.figure(figsize=(10, 6))
colors = ['skyblue' if i not in highlight_indices else 'coral' for i in range(len(sums))]
plt.barh(questions, sums, color=colors)

plt.title("Suma odpowiedzi dla każdego pytania")
plt.xlabel("Suma odpowiedzi")
plt.ylabel("Numery pytań")

plt.show()
```

**Zadanie 12**
Wyobraź sobie że masz siatkę 2D (dwu wymiarową tablice) reprezentującą temperatury na powierzchni prostokątnego materiału. Każda komórka siatki zawiera temperaturę w danym punkcie materiału. Wykorzystując bibliotekę `matplotlib`, zapoznaj się i wykorzystaj wykres [colorbar](https://matplotlib.org/stable/gallery/images_contours_and_fields/colormap_interactive_adjustment.html#sphx-glr-gallery-images-contours-and-fields-colormap-interactive-adjustment-py) do zaprezentowania symulacji rozchodzenia się ciepła.

Równanie przewodnictwa cieplnego w dwóch wymiarach może być uproszczone i przedstawione jako iteracyjny wzór:
$$ T^{(n+1)}_{i,j} = \frac{1}{4} (T^{(n)}_{i+1,j} + T^{(n)}_{i-1,j} + T^{(n)}_{i,j+1} + T^{(n)}_{i,j-1}) $$

Oznacza to, że temperatura w punkcie $T_{i,j}$ w kolejnym kroku czasu $n+1$ jest średnią z temperatury sąsiednich punktów w bieżącym kroku $n$. Brzegi siatki mają stałą temperaturę (warunki brzegowe).

Zaimplementuj model symulacji przepływu ciepła na siatce 2D o rozmiarze $40 \times 40$. Ustaw stałe wartości temperaturowe na górnym brzegu wynoszącą $100^{o}C$, a na dolnym brzegu: $0^{o}C$. Stwórz symulację dla $500$ kroków, przedstaw rozkład temperatury materiału na wykresie co $50$-ty krok symulacji.

Pamiętaj aby opisać każdy wykres w odpowiednie informacje

```python
from matplotlib import cm

N = 40
num_steps = 500
T = np.zeros((N, N))
T[0, :] = 100
T[-1, :] = 0

def heat_diffusion(T):
    T_new = T.copy()
    for i in range(1, N-1):
        for j in range(1, N-1):
            T_new[i, j] = 0.25 * (T[i+1, j] + T[i-1, j] + T[i, j+1] + T[i, j-1])
    return T_new

for step in range(1, num_steps + 1):
    T = heat_diffusion(T)
    if step % 50 == 0:
        plt.figure(figsize=(6, 6))
        plt.imshow(T, cmap=cm.coolwarm, interpolation='nearest')
        plt.colorbar(label='Temperatura (°C)')
        plt.title(f'Symulacja przepływu ciepła - Krok {step}')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.show()
```
