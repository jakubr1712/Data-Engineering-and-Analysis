**Zadanie 7**

Wyznacz podstawowe statystyki próbkowe (średnia, odchylenie standardowe, kwantyle) dla liczby pasażerów (`flights`) w każdym roku z osobna. Przedstaw wyniki w tak sposób, by zagregowane wartości przechowywane były w kolumnach, a kolejne lata - w wierszach.

````python
import pandas as pd

stats = flights.groupby('year')['passengers'].agg(
    mean='mean',
    std='std',
    q25=lambda x: x.quantile(0.25),
    median=lambda x: x.median(),
    q75=lambda x: x.quantile(0.75)
).reset_index()

stats.set_index('year', inplace = True)

print(stats)
````

**Zadanie 8**
Pod koniec semestru studenci piszą test zalieczeniowy z którego można uzyskać $100$ punktów. Podzielmy możliwe do uzyskania wyniki punktowe z testu na $10$ grup z godnie ze słownikiem:
```python
grupy = {
    1: (0, 10),   
    2: (11, 20),
    3: (21, 30),   
    4: (31, 40),  
    5: (41, 50),  
    6: (51, 60),   
    7: (61, 70),   
    8: (71, 80),  
    9: (81, 90),   
    10: (91, 100)  
}
```
Z wieloletnich obserwacji wynika, że prawdopodobieństwa uzyskania wyniku z danej grupy punktowej są następujące:
```python
prawdopodobienstwa = [0.02, 0.03, 0.05, 0.1, 0.13, 0.16, 0.22, 0.18, 0.08, 0.03]
```
dla kolejnych grup punktowych.

Na podstawie otrzymanych danych zbuduj przykładdowe dane (ramkę danych) wyników z testu zaliczeniowego dla $200$ studentów (ustal ziarno generatora pseudolosowego).

Następnie, zgodnie ze skalą akadmicką wykorzystując fukcję `cut()` punkty - zmienną ilościową na zmieną kategoryczną:
* grupy 1-5 - ndst,
* grupa 6 - dst,
* grupa 7 - dst+,
* grupa 8 - db,
* grupa 9 - db+,
* grupa 10 - bdb.

````python
import numpy as np


groups = {
    1: (0, 10),
    2: (11, 20),
    3: (21, 30),
    4: (31, 40),
    5: (41, 50),
    6: (51, 60),
    7: (61, 70),
    8: (71, 80),
    9: (81, 90),
    10: (91, 100)
}

probabilities = [0.02, 0.03, 0.05, 0.1, 0.13, 0.16, 0.22, 0.18, 0.08, 0.03]

results = np.random.choice(
    list(groups.keys()),
    size=200,
    p=probabilities
)

points = []
for group in results:
    points.append(np.random.randint(groups[group][0], groups[group][1] + 1))

df = pd.DataFrame({'Student': range(1, 201), 'Points': points})

bins = [-1, 51, 61, 71, 81, 91, 101]
labels = ['ndst', 'dst', 'dst+', 'db', 'db+', 'bdb']

df['Grades'] = pd.cut(df['Points'], bins=bins, labels=labels)

print(df)

````

````python
import pandas as pd
szczepionki_df = pd.read_csv('/content/drive/MyDrive/Analiza_Danych/szczepionki.csv', delimiter=';')
wizyty_df = pd.read_csv('/content/drive/MyDrive/Analiza_Danych/wizyty.csv', delimiter=';')

print("----Podpunkt 1----")

szczepionki_wieksze_niz_5 = szczepionki_df[szczepionki_df['liczba_dawek'] > 5]
if szczepionki_wieksze_niz_5.empty:
    print("Nie ma szczepionek, które mają rekomendowane więcej niż 5 dawek.")
else:
    print(f"Szczepionki, dla których rekomendowana liczba dawek jest większa niż 5:\n{szczepionki_wieksze_niz_5.to_string(index=False)}")

wizyty_df['data_szczepienia'] = pd.to_datetime(wizyty_df['data_szczepienia'])
szczepienia_lipiec_sierpien_2023 = wizyty_df[(wizyty_df['data_szczepienia'].dt.year == 2023) & (wizyty_df['data_szczepienia'].dt.month.isin([7, 8]))]
liczba_szczepien_lipiec_sierpien_2023 = szczepienia_lipiec_sierpien_2023.shape[0]
print(f"Liczba szczepień w lipcu i sierpniu 2023 roku: {liczba_szczepien_lipiec_sierpien_2023}")

print("----Podpunkt 2----")

dawki_per_szczepionka = wizyty_df.groupby('kod_szczepionki').size().reset_index(name='liczba_dawk')

dawki_per_szczepionka_sorted = dawki_per_szczepionka.sort_values(by='liczba_dawk', ascending=False)

print(dawki_per_szczepionka_sorted.to_string(index=False))

print("----Podpunkt 3----")

pacjenci_sz12_3d = wizyty_df[wizyty_df['kod_szczepionki'] == 'sz12_3d']

unikalni_pacjenci_sz12_3d = pacjenci_sz12_3d['pesel'].unique()
liczba_pacjentow_sz12_3d = len(unikalni_pacjenci_sz12_3d)

kobiety_sz12_3d = [pesel for pesel in unikalni_pacjenci_sz12_3d if int(str(pesel)[-2]) % 2 == 0]
liczba_kobiet_sz12_3d = len(kobiety_sz12_3d)

print(f"Liczba różnych pacjentów, którzy przyjęli przynajmniej jedną dawkę szczepionki sz12_3d: {liczba_pacjentow_sz12_3d}")
print(f"Liczba kobiet wśród tych pacjentów: {liczba_kobiet_sz12_3d}")
````

4. Wyszukaj 5 produktów o najniższej cenie i zasymuluj jak by wyglądały wyniki sprzedaży jeżeli produkty byłyby sprzedawane po wyższej cenie. Wykonaj symulację dla cen pomnożonych przez `1.2`, `1.5`, `1.7` wyniki zapisz w osobnych plikach *.csv*


````python
df = pd.read_csv('/content/drive/MyDrive/Analiza_Danych/products.csv')


df['total_sales'] = df['price'] * df['quantity']

sorted_df = df.sort_values(by='total_sales')
lowest_sales = sorted_df.head(3)
highest_sales = sorted_df.tail(3)

print("Three products with the lowest total sales:")
print(lowest_sales[['product_name', 'total_sales']])

print("\nThree products with the highest total sales:")
print(highest_sales[['product_name', 'total_sales']])


average_price = df['price'].mean()
price_range = df[(df['price'] >= average_price - 50) & (df['price'] <= average_price + 50)]

highest_sales_product = sorted_df.tail(1)

print("\nProducts within the price range of average price ± 50:")
print(price_range[['product_name', 'price']])

print("\nProduct with the highest total sales:")
print(highest_sales_product[['product_name', 'total_sales']])


lowest_price_products = df.nsmallest(5, 'price')
simulated_sales_1_2 = lowest_price_products.copy()
simulated_sales_1_5 = lowest_price_products.copy()
simulated_sales_1_7 = lowest_price_products.copy()


simulated_sales_1_2['price'] *= 1.2
simulated_sales_1_2['total_sales'] = simulated_sales_1_2['price'] * simulated_sales_1_2['quantity']

simulated_sales_1_5['price'] *= 1.5
simulated_sales_1_5['total_sales'] = simulated_sales_1_5['price'] * simulated_sales_1_5['quantity']

simulated_sales_1_7['price'] *= 1.7
simulated_sales_1_7['total_sales'] = simulated_sales_1_7['price'] * simulated_sales_1_7['quantity']

simulated_sales_1_2.to_csv('simulated_sales_1_2.csv', index=False)
simulated_sales_1_5.to_csv('simulated_sales_1_5.csv', index=False)
simulated_sales_1_7.to_csv('simulated_sales_1_7.csv', index=False)
````
